1.CALCULATE THE NUMBER OF MALE AND FEMALE PATIENTS?

```python
males=df[df["Gender"]=="Male"]["Age"].count()
females=df[df["Gender"]=="Female"]["Age"].count()
print(f"Number of males:{males}")
print(f"Number of females:{females}")
df["Gender"].max()
total=df.groupby("Gender")["Age"].sum()
plt.pie(total,labels=total.index,autopct="%1.1f%%",startangle=90)
plt.axis("equal")
plt.show()
```

6.WHO ADMITTED MORE DAYS IN THE HOSPITAL?

```python
df["Date of Admission"]=pd.to_datetime(df["Date of Admission"])
df["Discharge Date"]=pd.to_datetime(df["Discharge Date"])
df["time_duration"]=df["Discharge Date"] - df["Date of Admission"]
longest_stay_patient=df.loc[df["time_duration"].idxmax()]
print(longest_stay_patient["Name"])
print("patient",longest_stay_patient,["time_duration"],"days")
```

14.IDENTIFY THE PERSONS ADMISSION DATE AND DISCHARGE DATE ?

```python
df["Date of Admission"]=pd.to_datetime(df["Date of Admission"])
df["Discharge Date"]=pd.to_datetime(df["Discharge Date"])
df["length"]=(df["Discharge Date"]-df["Date of Admission"]).dt.days
plt.figure(figsize=(10,6))
plt.bar(df["Date of Admission"],df["Age"])
plt.title("relation")
plt.xlabel("admission")
plt.ylabel("discharge")
plt.show()
```

15.IDENTIFY THE PATIENTS RATE IN EACH YEAR?

```python
df["Date of Admission"]=pd.to_datetime(df["Date of Admission"])
df["years"]=df["Date of Admission"].dt.year
patients=df["years"].value_counts()
print(patients)
plt.figure(figsize=(6,4))
patients.plot(kind="bar")
plt.title("NUMBER OF PATIENTS PER YEAR")
plt.xlabel("YEAR")
plt.ylabel("NUMBER OF PATIENTS")
plt.show()
```

16.IDENTIFY THE DISEASES RATE IN EACH YEAR?

```python
df["Date of Admission"]=pd.to_datetime(df["Date of Admission"])
df["years"]=df["Date of Admission"].dt.year
disease_rate=df.groupby(["years","Medical Condition"]).size().unstack("Medical Condition")
print(disease_rate)
plt.figure(figsize=(8,6))
disease_rate.plot(kind="bar")
plt.title("NUMBER OF PATIENTS PER YEAR")
plt.xlabel("YEAR")
plt.ylabel("NUMBER OF PATIENTS")
plt.show()
```

#CONCLUSION

The dataset provides a comprehensive overview of healthcare trends through key questions. It reveals an almost equal gender distribution among patients, with males and females accounting for 27,774 and 27,726 individuals, respectively. Chronic diseases dominate, with Arthritis being the most common condition overall, particularly among females, while Diabetes is prevalent among males. Young patients under 30 years make up 9,693 cases, highlighting significant healthcare needs in this group. The longest hospital stay recorded was 30 days for a Diabetes patient, emphasizing the burden of chronic illnesses. Admission types (urgent, elective, and emergency) indicate varying healthcare demands, while cancer patients average age and billing amounts provide insight into treatment costs and patient demographics. Medications prescribed, test results, and blood group distributions add further detail about patient care. Yearly trends in patient rates and disease prevalence help track healthcare demand over time. These findings collectively offer valuable insights for hospital resource planning and public health strategies.
